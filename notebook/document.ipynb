```python
##create sample text file
import os
os.makedirs("../data/text_files",exist_ok=True)
```

```python
sample_texts={
    "../data/text_files/python_intro.txt":"""Python is a high-level, interpreted programming language known for its simplicity, readability, and versatility. Created by Guido van Rossum and first released in 1991, Python was designed with an emphasis on clean, easy-to-understand syntax that allows programmers to express concepts in fewer lines of code compared to many other languages. Its straightforward structure makes it an excellent choice for beginners, while its powerful features appeal to experienced developers as well.

One of Python’s greatest strengths is its wide range of applications. It is used in web development, data science, machine learning, automation, artificial intelligence, scientific computing, and more. Python offers a rich ecosystem of libraries such as NumPy, Pandas, TensorFlow, Flask, and Django, which enable developers to build anything from simple scripts to complex, large-scale applications. This extensive support has helped Python become one of the most popular and widely used programming languages in the world.

Another key advantage of Python is its strong and active community. Developers around the globe contribute to its continuous improvement by creating libraries, frameworks, and tools that expand Python’s capabilities. The language also has comprehensive documentation and countless learning resources, making it easier for new learners to get started. Python’s open-source nature ensures it remains freely available to everyone.

Overall, Python’s combination of simplicity, flexibility, and robust community support makes it an ideal choice for both beginners exploring programming for the first time and professionals building advanced applications. Its consistent growth and increasing demand across industries highlight its importance in the modern technological landscape."""
}

for filepath,content in sample_texts.items():
    with open(filepath,'w',encoding="utf-8") as f:
        f.write(content)

print("sample text files created")
```

```python
##TextLoader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")
document = loader.load()
print(document)
```

```python
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={'encoding':'utf-8'},
    show_progress=False
)
documents = dir_loader.load()
print(documents)
```

```python
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False
)
pdf_documents = dir_loader.load()
pdf_documents
```
